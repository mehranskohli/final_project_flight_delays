```python
data_sklearn = data.copy()

from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
data_sklearn['Origin_code'] = lb_make.fit_transform(data['AirportFrom'])
data_sklearn['Airline_code'] = lb_make.fit_transform(data['Airline'])


data_sklearn.head()
```

```python
data_sklearn.dtypes
```

```python
#data_dummy = pd.get_dummies(data=length_data,drop_first=True)
#data_dummy.head()
```

```python
data_y = data_sklearn.Delay
data_x = data_sklearn[["Length", "DayOfWeek", "Origin_code", "Airline_code" ]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2)
```

```python
# Create Decision Tree classifer object
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier

clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

print("DecisionTreeClassifier Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
y_pred_proba = clf.predict_proba(X_test)[::,1]
fpr, tpr, dummy = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr, color='b')
print('Area Under Curve:', auc)
```

```python
clf.feature_importances_
#Length of Flights has the highest importance
```

```python
from sklearn.ensemble import GradientBoostingClassifier

clf1 = GradientBoostingClassifier()
clf1 = clf1.fit(X_train,y_train)
y_pred1 = clf1.predict(X_test)
```
